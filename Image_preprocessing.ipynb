```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

PATH='D:\\Code\\kural_work\\movie_generation\\data\\'
img_ = Image.open(PATH+'s01_5m_2s_100ms_25p.tif')
nf = img_.n_frames #number of frames
```

```python
d0, d1 = 230, 245
rad = 130
img_.seek(nf-1)
full_img = np.asarray(list(img_.getdata())).reshape(512,512)
plt.imshow(full_img[d0-rad:d0+rad,d1-rad:d1+rad])
plt.show()
```

```python
num_imgs = 15
center = [230, 245]
edge_param = 0.7
r = rad-edge_param*128//2
crop_cent = []
for i in range(num_imgs):
    crop_cent.append((np.int(center[0]+r*np.sin(2*np.pi*i/num_imgs)), np.int(center[1]+r*np.cos(2*np.pi*i/num_imgs))))
```

```python
width = 128
for d0,d1 in crop_cent:
    plt.imshow(full_img[d0-width//2:d0+width//2,d1-width//2:d1+width//2])
    plt.show()
```
