```python
file_list = []
for dirpath, dirnames, filenames in os.walk(PATH):
    for file in filenames:
        if '.nd2' in file:
            file_list.append(os.path.join(dirpath, file)[len(PATH)+1:])
```

```python
full_path_list = []
for dirpath, dirnames, filenames in os.walk(PATH):
    for file in filenames:
        if '.nd2' in file:
            full_path_list.append(os.path.join(dirpath, file))
```

```python
test_file = np.random.choice(full_path_list)
print(test_file)
img_file = ND2Reader(test_file)
im = img_file.metadata
h = im['height']
w = im['width']
nf = len(im['frames'])
nz = len(im['z_levels'])
nc = len(im['channels'])
ps = im['pixel_microns']

print(f"height = {h}\nwidth = {w}\nnumber of frames = {nf}\nnumber of z stacks = {nz}\nnumber of channels = {nc}\npixel size = {ps}")
```

```python
data_dict = {'name':[], 'height':[], 'width':[], 'n_frames':[], 'n_stacks':[], 'n_channels':[], 'pixel_size':[]}
for test_file in full_path_list:
    print(test_file)
    data_dict['name'].append(test_file)
    img_file = ND2Reader(test_file)
    im = img_file.metadata
    data_dict['height'].append(im['height'])
    data_dict['width'].append(im['width'])
    data_dict['n_frames'].append(len(im['frames']))
    data_dict['n_stacks'].append(len(im['z_levels']))
    data_dict['n_channels'].append(len(im['channels']))
    data_dict['pixel_size'].append(im['pixel_microns'])
```
